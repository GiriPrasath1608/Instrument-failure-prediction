```python
import pandas as pd
data = pd.read_csv('fp_csv.csv')
df = pd.DataFrame(data)
```

```python
df = df.drop(columns = ['Unnamed: 0'],axis = 1)
```

```python
df['Difference Temperature'] = df['Difference Temperature'].round(2)
```

```python
data = []
for i in range(df.shape[0]):
    data1 = {}
    data1['UDI']                      = int(df['UDI'][i])
    data1['Product ID']               = df['Product ID'][i]
    data1['Type']                     = df['Type'][i]
    data1['Air temperature [K]']      = float(df['Air temperature [K]'][i])
    data1['Process temperature [K]']  = float(df['Process temperature [K]'][i])
    data1['Rotational speed [rpm]']   = int(df['Rotational speed [rpm]'][i])
    data1['Torque [Nm]']              = float(df['Torque [Nm]'][i])
    data1['Tool wear [min]']          = int(df['Tool wear [min]'][i])
    data1['Target']                   = int(df['Target'][i])
    data1['Failure Type']             = df['Failure Type'][i]
    data1['Target']                   = int(df['Target'][i])
    data1['Noise Power']              = float(df['Noise Power'][i])
    data1['Noise']                    = float(df['Noise'][i])
    data1['Difference Temperature']   = float(df['Difference Temperature'][i])
    data.append(data1)
data
```
