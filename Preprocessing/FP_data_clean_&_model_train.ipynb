# **2. FEATURE ENGINEERING**

● rotational speed [rpm]: calculated from power of 2860 W, overlaid with a normally
distributed noise

```python
# noise power
import math
rpm = df['Rotational speed [rpm]'][9999]
torque = df['Torque [Nm]'][9999]
power = (2*math.pi*rpm*torque)/60          # out put power calculation formula = (2*pi*N*T)/60
power
```

```python
df['Power'] = 2860             # creating  power column
```

```python
for i in range(df.shape[0]):
  rpm = df['Rotational speed [rpm]'][i]
  torque = df['Torque [Nm]'][i]
  df.loc[i,'Noise Power'] = (2*math.pi*rpm*torque)/60       # creating
```

```python
df['Noise'] = df['Noise Power'] - df['Power']
```

```python
df.head()
```

```python
df['Difference Temperature'] = (df['Process temperature [K]'] - df['Air temperature [K]'])/2
```

```python
df.head(5)
```

```python
df = df.drop(columns = ['Power'], axis = 1)
```

```python
df.head(5)
```

```python
df.head(5)
```

```python
df[['Noise Power','Noise']] = df[['Noise Power','Noise']].round(2)
df['Rotation per torque'] = df['Rotational speed [rpm]']/df['Torque [Nm]']
df['Torque per Rotation'] = df['Torque [Nm]']/df['Rotational speed [rpm]']
df['Tool wear per Rotation'] = df['Tool wear [min]']/df['Rotational speed [rpm]']
df['Tool wear per torque'] = df['Tool wear [min]']/df['Torque [Nm]']
df['Tool wear per power'] = df['Tool wear [min]']/df['Noise Power']
```

```python
label_encoder = LabelEncoder()

X = df.drop(columns = ['Target','Failure Type','Product ID','UDI','Unnamed: 0'], axis=1)
X['Type'] = LabelEncoder().fit_transform(df['Type'])

y = df['Target']

X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 0.20, random_state=100,stratify = y)
print(len(X_train),len(X_test),len(y_train),len(y_test))
```

## **MODEL SELECTION**

```python
from imblearn.over_sampling import SMOTE
evaluation_test = []
evaluation_train =[]
#logistic model
for model in models:

    print('model name =', model)
    #model train
    X_sampled,y_sampled = SMOTE().fit_resample(X_train,y_train)
    model.fit(X_sampled,y_sampled)

    # model prediction
    y_pred_test = model.predict(X_test)

    #model evaluation on test data
    test_accuracy = accuracy_score(y_test,y_pred_test)
    test_precision_score = precision_score(y_test,y_pred_test)
    test_recall_score = recall_score(y_test,y_pred_test)
    test_f1_score = f1_score(y_test,y_pred_test)
    test_conf_matrix = confusion_matrix(y_test,y_pred_test)
    tn, fp, fn, tp = confusion_matrix(y_test,y_pred_test).ravel()
    specificity_test = (tn/(fp+tn))

    print('Testing Evaluation--------------------------------------------------------------------')
    print('test_accuracy',test_accuracy)
    print('test_precision_score',test_precision_score)
    print('test_recall_score',test_recall_score)
    print('test_f1_score',test_f1_score)
    print('test_conf_matrix',test_conf_matrix)
    print('specificity_test: ',specificity_test)

    # Plot ROC curve
    y_prob_test = model.predict_proba(X_test)[:,1]
    fpr,tpr,thershold = roc_curve(y_test, y_prob_test)
    roc_auc = auc(fpr, tpr)

    plt.figure()
    plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic')
    plt.legend(loc="lower right")
    print('roc_auc',roc_auc)
    plt.show()

    data_test={'model name': model,
              'test_accuracy':test_accuracy,
              'test_precision_score':test_precision_score,
              'test_recall_score':test_recall_score,
              'test_f1_score':test_f1_score,
              'test_conf_matrix':str(test_conf_matrix),
              'roc_auc':roc_auc}

    evaluation_test.append(data_test)

    #model evaluation in train data
    y_pred_train = model.predict(X_train)

    train_accuracy = accuracy_score(y_train,y_pred_train)
    train_precision_score = precision_score(y_train,y_pred_train)
    train_recall_score = recall_score(y_train,y_pred_train)
    train_f1_score = f1_score(y_train,y_pred_train)
    train_conf_matrix = confusion_matrix(y_train,y_pred_train)
    tn, fp, fn, tp = confusion_matrix(y_train,y_pred_train).ravel()
    specificity_train = (tn/(fp+tn))

    print('Training Evaluation--------------------------------------------------------------------')
    print('train_accuracy',train_accuracy)
    print('train_precision_score',train_precision_score)
    print('train_recall_score',train_recall_score)
    print('train_f1_score',train_f1_score)
    print('train_conf_matrix',train_conf_matrix)
    print('specificity_train: ',specificity_train)

    # Plot ROC curve
    y_prob_train = model.predict_proba(X_train)[:,1]
    fpr,tpr,thershold = roc_curve(y_train, y_prob_train)
    roc_auc_train = auc(fpr, tpr)

    plt.figure()
    plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc_train)
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic')
    plt.legend(loc="lower right")
    print('roc_auc_train',roc_auc_train)
    plt.show()
    data_train = {'model name' : model,
                'train_accuracy':train_accuracy,
                'train_precision_score':train_precision_score,
                'train_recall_score':train_recall_score,
                'train_f1_score':train_f1_score,
                'train_conf_matrix':str(train_conf_matrix),
                'train_roc_auc':roc_auc_train}
    evaluation_train.append(data_train)
```

## **HYPER PARAMETER TUNNING**

```python
# Base estimator
base_estimator = DecisionTreeClassifier()

# AdaBoost model
ada_model = AdaBoostClassifier(base_estimator=base_estimator)

param_grid = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 0.5, 1],
    'base_estimator__max_depth': [1, 2, 3, 4, 5],
    'base_estimator__min_samples_split': [2, 5, 10],
    'base_estimator__min_samples_leaf': [1, 2, 4],
}

random_search = RandomizedSearchCV(
    estimator=ada_model,
    param_distributions=param_grid,
    n_iter=20,  # Number of random combinations to try
    scoring='recall',  # Assuming recall is important
    cv=5,
    n_jobs=-1,  # Use all available cores
    verbose=2,
    random_state=42
)

X_sampled,y_sampled = SMOTE().fit_resample(X_train,y_train)
random_search.fit(X_sampled,y_sampled)
model = random_search.best_estimator_
y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test,y_pred)
precision = precision_score(y_test,y_pred)
recall = recall_score(y_test,y_pred)
f1 = f1_score(y_test,y_pred)
conf_matrix = confusion_matrix(y_test,y_pred)
tn, fp, fn, tp = confusion_matrix(y_test,y_pred).ravel()
specificity = tn/(fp+tn)

print('accuracy = ',accuracy)
print('precision = ',precision)
print('recall = ',recall)
print('specificity = ',specificity)
print('f1 = ',f1)
print('conf_matrix = ',conf_matrix)

y_proba = model.predict_proba(X_test)[:,1]
fpr,tpr,thershold = roc_curve(y_test, y_proba)
auc_score = auc(fpr,tpr)

print('auc_score = ',auc_score)
```
