```python
logistic_model = LogisticRegression()
K_neighbor_model = KNeighborsClassifier()
decision_tree_model = DecisionTreeClassifier()
random_forest_model = RandomForestClassifier()
gbc_model = GradientBoostingClassifier()
adaboost_model = AdaBoostClassifier(estimator = DecisionTreeClassifier(), n_estimators = 200,learning_rate=1)
models = [logistic_model,K_neighbor_model,decision_tree_model,random_forest_model,gbc_model,adaboost_model]
```

```python
label_encoder = LabelEncoder()
scaler = StandardScaler()

X = df.drop(columns = ['Target','Failure Type','Product ID','UDI','Unnamed: 0'], axis=1)
X['Rotational speed [rpm]'] = np.log(df['Rotational speed [rpm]'])
X['Type'] = LabelEncoder().fit_transform(df['Type'])

y = df['Target']

X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 0.20, random_state=100,stratify = y)
print(len(X_train),len(X_test),len(y_train),len(y_test))
```

```python
from imblearn.over_sampling import SMOTE
evaluation_test = []
evaluation_train =[]
#logistic model
for model in models:

    print('model name =', model)
    #model train
    X_sampled,y_sampled = SMOTE().fit_resample(X_train,y_train)
    model.fit(X_sampled,y_sampled)

    # model prediction
    y_pred_test = model.predict(X_test)

    #model evaluation on test data
    test_accuracy = accuracy_score(y_test,y_pred_test)
    test_precision_score = precision_score(y_test,y_pred_test)
    test_recall_score = recall_score(y_test,y_pred_test)
    test_f1_score = f1_score(y_test,y_pred_test)
    test_conf_matrix = confusion_matrix(y_test,y_pred_test)
    tn, fp, fn, tp = confusion_matrix(y_test,y_pred_test).ravel()
    specificity_test = (tn/(fp+tn))

    print('Testing Evaluation--------------------------------------------------------------------')
    print('test_accuracy',test_accuracy)
    print('test_precision_score',test_precision_score)
    print('test_recall_score',test_recall_score)
    print('test_f1_score',test_f1_score)
    print('test_conf_matrix',test_conf_matrix)
    print('specificity_test: ',specificity_test)

    # Plot ROC curve
    y_prob_test = model.predict_proba(X_test)[:,1]
    fpr,tpr,thershold = roc_curve(y_test, y_prob_test)
    roc_auc = auc(fpr, tpr)

    plt.figure()
    plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic')
    plt.legend(loc="lower right")
    print('roc_auc',roc_auc)
    plt.show()

    data_test={'model name': model,
              'test_accuracy':test_accuracy,
              'test_precision_score':test_precision_score,
              'test_recall_score':test_recall_score,
              'test_f1_score':test_f1_score,
              'test_conf_matrix':str(test_conf_matrix),
              'roc_auc':roc_auc}

    evaluation_test.append(data_test)

    #model evaluation in train data
    y_pred_train = model.predict(X_train)

    train_accuracy = accuracy_score(y_train,y_pred_train)
    train_precision_score = precision_score(y_train,y_pred_train)
    train_recall_score = recall_score(y_train,y_pred_train)
    train_f1_score = f1_score(y_train,y_pred_train)
    train_conf_matrix = confusion_matrix(y_train,y_pred_train)
    tn, fp, fn, tp = confusion_matrix(y_train,y_pred_train).ravel()
    specificity_train = (tn/(fp+tn))

    print('Training Evaluation--------------------------------------------------------------------')
    print('train_accuracy',train_accuracy)
    print('train_precision_score',train_precision_score)
    print('train_recall_score',train_recall_score)
    print('train_f1_score',train_f1_score)
    print('train_conf_matrix',train_conf_matrix)
    print('specificity_train: ',specificity_train)

    # Plot ROC curve
    y_prob_train = model.predict_proba(X_train)[:,1]
    fpr,tpr,thershold = roc_curve(y_train, y_prob_train)
    roc_auc_train = auc(fpr, tpr)

    plt.figure()
    plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc_train)
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic')
    plt.legend(loc="lower right")
    print('roc_auc_train',roc_auc_train)
    plt.show()
    data_train = {'model name' : model,
                'train_accuracy':train_accuracy,
                'train_precision_score':train_precision_score,
                'train_recall_score':train_recall_score,
                'train_f1_score':train_f1_score,
                'train_conf_matrix':str(train_conf_matrix),
                'train_roc_auc':roc_auc_train}
    evaluation_train.append(data_train)
```
